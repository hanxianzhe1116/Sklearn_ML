```python
Xtrain,Xtest,Ytrain,Ytest = TTS(data_325[ new_feature_names ], label, test_size=0.3, random_state=111)
```

```python
RON_sun_train, RON_sun_test = Xtrain['RON损失'], Xtest['RON损失']
```

```python
new_Xtrain = Xtrain.drop(['RON损失'], axis=1)
new_Xtest = Xtest.drop(['RON损失'], axis=1)
```

```python
#Xtrain,Xtest,Ytrain,Ytest = TTS(X, y, test_size=0.3, random_state=111)
for i in range(1,1001):
    forest_reg = RandomForestRegressor(n_estimators=i)
    forest_reg.fit(new_Xtrain, Ytrain)
    y_predict = forest_reg.predict(new_Xtest)
    mse_score = MSE(Ytest,y_predict)
    rmse_score = np.sqrt(mse_score)
    mae_score = mean_absolute_error(Ytest,y_predict)
    rr_score = r2_score(Ytest,y_predict)

    print(i, mse_score, rmse_score, mae_score, rr_score)
```

```python
new_Xtest.columns
```

```python
RON = new_Xtest['原料性质：辛烷值'] - y_predict
```

```python
RON
```

```python
mse_score = MSE(RON_sun_test,RON)
rmse_score = np.sqrt(mse_score)
mae_score = mean_absolute_error(RON_sun_test,RON)
rr_score = r2_score(RON_sun_test,RON)

print(mse_score, rmse_score, mae_score, rr_score)
```
